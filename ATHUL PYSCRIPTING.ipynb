## Define get_all_products(data)

### Subtask:

**Reasoning**:
I need to define the `get_all_products` function as described, iterating through the nested dictionary to collect all product dictionaries into a single list.



```python
def get_all_products(data):
    all_products = []
    for category_data in data["categories"].values():
        all_products.extend(category_data["products"])
    return all_products

# Test the function with the existing inventory data
all_inventory_products = get_all_products(inventory)

print("All products in a flat list:")
display(all_inventory_products)
```

### Subtask:
Create a function `get_total_stock(cat_name)` that takes a category name as input and returns the total stock count for all products within that specific category.


### Reasoning:
To calculate the total stock for a specific category, I will define a function `get_total_stock` that takes the inventory data and the category name as input. ## Define get_avg_price()

### Subtask:
Create a function `get_avg_price()` that computes the inventory-wide average product price. This function should use `map()` to extract prices and `functools.reduce()` to sum them, then divide by the total number of products. Make sure to import `functools`.


### Reasoning:
To calculate the inventory-wide average product price, I will first get a flat list of all products. Then, I will use `map()` to extract the 'price' of each product. Finally, I will use `functools.reduce()` to sum these prices and divide by the total number of products to get the average. I will need to import the `functools` module for `reduce`.

**Reasoning**:
Now I will implement the `get_avg_price` function using `map()` and `functools.reduce()` as described, and test it to compute the average product price.



```python
import functools

def get_avg_price(data):
    all_products = get_all_products(data)
    if not all_products:
        return 0.0

    # Extract prices using map
    prices = list(map(lambda product: product['price'], all_products))

    # Sum prices using functools.reduce
    total_price = functools.reduce(lambda x, y: x + y, prices)

    # Calculate average
    average_price = total_price / len(all_products)
    return average_price

# Test the function
avg_price = get_avg_price(inventory)
print(f"\nInventory-wide average product price: {avg_price:.2f}")
```

## Task 3.1: Create a custom iterator class ProductIterator.

### Subtask:
Define a Python class `ProductIterator` that implements the iterator protocol (`__iter__` and `__next__`) to iterate through and return the names of all products in the `inventory` one by one.


### Reasoning:
To create the `ProductIterator` class, I will first define its `__init__` method to flatten the nested `inventory` dictionary into a list of all product names and initialize a pointer for iteration. Then, I will implement `__iter__` to return the instance itself, making it an iterator. Finally, the `__next__` method will sequentially return product names, raising `StopIteration` when all products have been yielded, adhering to Python's iterator protocol.

**Reasoning**:
Now I will implement the `ProductIterator` class as described, including its `__init__`, `__iter__`, and `__next__` methods, and then test it by iterating through an instance to print all product names.



```python
class ProductIterator:
    def __init__(self, data):
        self._product_names = []
        for category_data in data["categories"].values():
            for product in category_data["products"]:
                self._product_names.append(product["name"])
        self._index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self._index < len(self._product_names):
            product_name = self._product_names[self._index]
            self._index += 1
            return product_name
        else:
            raise StopIteration

# Create an instance of ProductIterator
product_iterator = ProductIterator(inventory)

print("\nIterating through product names using ProductIterator:")
for name in product_iterator:
    print(f"- {name}")
```

## Task 3.2: Build a generator function supplier_generator().

### Subtask:
Create a generator function `supplier_generator()` that iterates through all products in the `inventory` and yields each unique supplier name associated with any product.


### Reasoning:
To create the `supplier_generator` function, I will iterate through all categories and their products in the `inventory`. For each product, I will go through its list of suppliers. To ensure that only unique supplier names are yielded, I will maintain a `seen_suppliers` set. Before yielding a supplier name, I will check if it's already in this set; if not, I will add it to the set and then yield it.

**Reasoning**:
Now I will implement the `supplier_generator` function as described, including its definition and a loop to iterate through it, demonstrating its functionality by printing each unique supplier name.



```python
def supplier_generator(data):
    seen_suppliers = set()
    for category_data in data["categories"].values():
        for product in category_data["products"]:
            for supplier in product["suppliers"]:
                if supplier not in seen_suppliers:
                    seen_suppliers.add(supplier)
                    yield supplier

print("\nUnique Supplier Names using supplier_generator:")
for supplier_name in supplier_generator(inventory):
    print(f"- {supplier_name}")
```

Construct a three-step generator pipeline: a generator to get all product dictionaries; a generator to filter products with price > 50.00; and a final generator yielding `(name, price)` tuples.

## Task 6.1: Build a closure `make_discount_calculator(rate)`.

### Subtask:
Create a closure `make_discount_calculator(rate)` that applies a given discount rate to an `original_price`.


### Reasoning:
To create the `make_discount_calculator` closure, I will define an outer function that accepts a `rate` argument. Inside this outer function, I will define a nested function, `calculate_discounted_price`, which will capture the `rate` from its enclosing scope. This nested function will take an `original_price` and compute the discounted price using the captured `rate`. The outer function will then return this nested function. Finally, I will demonstrate the closure by creating two separate discount calculators with different rates and applying them to an example price to verify their functionality.

**Reasoning**:
Now I will implement the `make_discount_calculator` function as described, including the nested function and the demonstration of its usage with different discount rates and an example price.



```python
def make_discount_calculator(rate):
    """Creates a closure that applies a given discount rate to an original price."""
    def calculate_discounted_price(original_price):
        return original_price - (original_price * rate)
    return calculate_discounted_price

# Test the closure
# Create a 10% discount calculator
discount_10_percent = make_discount_calculator(0.10)

# Create a 25% discount calculator
discount_25_percent = make_discount_calculator(0.25)

example_price_1 = 100.00
example_price_2 = 50.00

print(f"\nOriginal price: ${example_price_1:.2f}")
print(f"Price after 10% discount: ${discount_10_percent(example_price_1):.2f}")
print(f"Price after 25% discount: ${discount_25_percent(example_price_1):.2f}")

print(f"\nOriginal price: ${example_price_2:.2f}")
print(f"Price after 10% discount: ${discount_10_percent(example_price_2):.2f}")
print(f"Price after 25% discount: ${discount_25_percent(example_price_2):.2f}")
```

### Reasoning:
To create the `create_category_filter` closure, I will define an outer function that accepts `cat_name`. Inside, I'll define a nested function `filter_products_by_category` that captures `cat_name`. This nested function will then take `all_products` as input and return a filtered list containing only products where `product['category']` matches the captured `cat_name`. The outer function will return this nested function. I will first define a helper function `get_all_products_with_category` to provide a list of products with their category information for the filter function to use. Finally, I will test the closure by creating a filter for 'Electronics' and apply it to an example list of all products.

**Reasoning**:
Before implementing the `create_category_filter` closure, I need a helper function that flattens the nested `inventory` and enriches each product dictionary with its `category` name. This is crucial because the filter function will need this category information to properly filter products.



```python
def get_all_products_with_category(data):
    all_products_categorized = []
    for category_name, category_data in data["categories"].items():
        for product in category_data["products"]:
            # Create a copy of the product and add the category name
            product_with_category = product.copy()
            product_with_category['category'] = category_name
            all_products_categorized.append(product_with_category)
    return all_products_categorized

# Test the helper function
products_with_categories = get_all_products_with_category(inventory)
print("\nAll products with their categories:")
for product in products_with_categories:
    print(f"- {product['name']} (Category: {product['category']})")
```

```python
def create_category_filter(cat_name):
    """Creates a closure that filters a list of products by a given category name."""
    def filter_products_by_category(all_products):
        return [product for product in all_products if product.get('category') == cat_name]
    return filter_products_by_category

# Test the closure
# Create a filter for 'Electronics' category
electronics_filter = create_category_filter('Electronics')

# Apply the filter to the list of all products with categories
electronics_products = electronics_filter(products_with_categories)

print(f"\nProducts in 'Electronics' category:")
for product in electronics_products:
    print(f"- {product['name']}")

# Create a filter for 'Clothing' category
clothing_filter = create_category_filter('Clothing')
clothing_products = clothing_filter(products_with_categories)

print(f"\nProducts in 'Clothing' category:")
for product in clothing_products:
    print(f"- {product['name']}")
```

## Task 6.5: Demonstrate value retention in closures with a simple counter.

### Subtask:
Create a closure `make_counter()` that returns a function. The returned function, when called, should increment and return a counter that is encapsulated within the closure's scope. This demonstrates how closures retain state between calls.

### Reasoning:
To demonstrate value retention, I will define an outer function `make_counter()` that initializes a `count` variable. Inside `make_counter()`, I will define a nested function `increment_counter()` that increments and returns this `count` variable. The key here is that `increment_counter()` will "remember" the `count` from its enclosing scope, even after `make_counter()` has finished executing. I will then create multiple counter instances to show that each maintains its own independent state.

**Reasoning**:
Now I will implement the `make_counter` closure as described, defining the outer function `make_counter` and the nested function `increment_counter`, and then demonstrate its usage by creating multiple counter instances and observing their independent state.



```python
def make_counter():
    """Creates a closure that encapsulates a counter and returns a function to increment it."""
    count = 0 # This count is encapsulated by the closure
    def increment_counter():
        nonlocal count # Declare count as non-local to modify it
        count += 1
        return count
    return increment_counter

# Test the closure
# Create two independent counter instances
counter1 = make_counter()
counter2 = make_counter()

print("\nDemonstrating independent counters:")

print(f"Counter 1: {counter1()}") # 1
print(f"Counter 1: {counter1()}") # 2
print(f"Counter 2: {counter2()}") # 1
print(f"Counter 1: {counter1()}") # 3
print(f"Counter 2: {counter2()}") # 2
```
